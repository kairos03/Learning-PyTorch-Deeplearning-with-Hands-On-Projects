```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0, 10, 100)
y1 = x1 ** 2
y2 = 10 * x1 + 2

x2 = 10 + 10 * np.random.randn(1000)
```

```python
plt.plot(x1, y1, 'r--', label='y1')
plt.plot(x1, y2, 'b-', label='y2')
plt.title("line plot")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc='upper left')
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x1, y1, 'g--', label='y1')
plt.subplot(1, 2, 2)
plt.plot(x1, y2, '-', label='y2')
plt.show()
```

```python
plt.hist(x2, 50, density=True, facecolor='orange', alpha=0.75)
plt.show()
```
